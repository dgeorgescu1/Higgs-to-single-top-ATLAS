## Weighting factors

```python
th_gamgam_shape = th_gamgam_sorted.shape[0]

ggf_shape = ggf_sorted.shape[0]
yyjets_shape = yyjets_sorted.shape[0]
ttH_shape = ttH_sorted.shape[0]
```

```python
###### Using Higgs Mass 125.09 GeV

Luminosity = 139e15 #b^-1

xsec_th_gamgam = 7.426e-14 #b

xsec_ttH = 5.065e-13 #b
xsec_th_gamgam = 7.426e-14 #b
xsec_ggf = 4.852e-11 #b
xsec_yyjj = 51.8e-12 #b

brf_Hyy = 2.27e-3

N_init = 4.86878e+08
```

```python
w_th_gamgam = Luminosity * xsec_th_gamgam * brf_Hyy

w_ggf = Luminosity * xsec_ggf * brf_Hyy
w_yyjets = Luminosity * xsec_yyjj * yyjets_shape/N_init
w_ttH = Luminosity * xsec_ttH * brf_Hyy

wghts_th_gamgam = w_th_gamgam * np.ones(th_gamgam_shape)
wghts_ggf = w_ggf * np.ones(ggf_shape)
wghts_yyjets = w_yyjets * np.ones(yyjets_shape)
wghts_ttH = w_ttH * np.ones(ttH_shape)
```

```python
cuts_y1_pt = np.arange(20000, 200000, 5000)
Props_y1_pt = []

for cut in (cuts_y1_pt):

    th_gamgam_cut = th_gamgam_sorted[th_gamgam_sorted.y1_pt > cut]
    
    ggf_cut = ggf_sorted[ggf_sorted.y1_pt > cut]
    yyjets_cut = yyjets_sorted[yyjets_sorted.y1_pt > cut]
    ttH_cut = ttH_sorted[ttH_sorted.y1_pt > cut]

    S = th_gamgam_cut.shape[0] * w_th_gamgam
    B1 = ggf_cut.shape[0] * w_ggf
    B2 = yyjets_cut.shape[0] * w_yyjets
    B3 = ttH_cut.shape[0] * w_ttH

    Prop = S / np.sqrt(S + B1 + B2 + B3)

    Props_y1_pt.append(Prop)
```

```python
def fom(x, obj):
    
    string = str(obj) + '_'
    E = x[string + 'E']
    px = abs(x[string + 'pt']) * np.cos(x[string + 'phi'])
    py = abs(x[string + 'pt']) * np.sin(x[string + 'phi'])
    pz = x[string + 'pt']/(np.tan(2*np.arctan(np.exp(-x[string + 'eta']))))

    return np.array([E, px, py, pz])
```

```python
datasets = [th_gamgam_sorted, ggf_sorted, yyjets_sorted, ttH_sorted]
```

```python
for data in datasets:
    
    #data['y1_fom'] = data.apply(lambda x: fom(x, 'y1'), axis = 1).apply(lambda x: x)
    #data['y2_fom'] = data.apply(lambda x: fom(x, 'y2'), axis = 1).apply(lambda x: x)
    
    for i in range(1,4):
        
        jet = 'jet' + str(i)
        data[jet + '_fom'] = data.apply(lambda x: fom(x, jet), axis = 1).apply(lambda x: x)
```

# Angle thing

```python
def eta(x):
    
    frac = x.wbos_pt/x.wbos_fom[3]
    
    if frac < 0: frac = -frac
    
    return - np.log(np.tan(0.5*np.arctan(frac)))
```

```python
def frac(x):
    
    D_phi = x.wbos_phi - x.jet1_phi
    D_eta = x.wbos_eta - x.jet1_eta
    
    return np.sqrt(D_phi**2 + D_eta**2)
```

```python
datasets = [th_gamgam_sorted, ggf_sorted, yyjets_sorted, ttH_sorted]
```

```python
for data in datasets:

    data['wbos_phi'] = data.apply(lambda x: 
                                      np.arctan(x.wbos_fom[2] / x.wbos_fom[1]) 
                                      ,axis = 1).apply(lambda x: x)

    data['wbos_pt'] = data.apply(lambda x:
                                    x.wbos_fom[1] / np.cos(x.wbos_phi),
                                     axis = 1).apply(lambda x: x)

    data['wbos_eta'] = data.apply(lambda x:
                                      eta(x),
                                      axis = 1).apply(lambda x: x)
    data['ang_wbos_jet1'] = data.apply(lambda x: frac(x) ,axis = 1).apply(lambda x: x)
```
